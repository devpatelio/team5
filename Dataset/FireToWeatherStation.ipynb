```python
dataset1 = df1.values
dataset1
#loc_dataset1 = dataset1[:,1:3]
#loc_dataset1
```

```python
dataset2 = df2.values
dataset2
```

```python
empty_array = np.empty([1633,2])
dataset1 = np.append(dataset1,empty_array, axis=1)
dataset1
```

```python
import math
closest_distance = 1000
closest_station = 0
for i in range(1633):
    for x in range(150):
        distance = math.sqrt(((dataset2[x,1]-dataset1[i,1]) ** 2)+((dataset2[x,2]-dataset1[i,2]) ** 2))
        if distance < closest_distance:
            closest_distance = distance
            closest_station = dataset2[x,0]
    closest_distance = closest_distance*69
    dataset1[i,5] = closest_distance
    dataset1[i,6] = closest_station
    closest_distance = 1000
```
